## Предобработка данных

### Исследование и заполнение пропусков в данных

Для поиска пропущенных значений обратимся к методу `isna()`, передав результат его работы методу `mean()`:

```python
#посчитаем долю пропусков и отсортируем результат по убыванию
round((data.isna().mean())*100, 1).sort_values(ascending=False)
```

Пропущенные значения присутствуют в следующих столбцах:
* `is_apartment` - здесь содержится больше всего пропусков, их доля практически 90%. Как правило, те, у кого апартаменты, знают об особенностях данного объекта недвижимости, поэтому пустое значение можно смело расценивать как False;
* `parks_nearest` - для заполнения этих пропусков возьмем медианные значения с группировкой по населенному пункту;
* `ponds_nearest` - для заполнения этих пропусков возьмем медианные значения с группировкой по населенному пункту;
* `balcony` - доля пропусков соствляет 48,6%. Цифра весьма существенна. Можно предположить, что за неимением балкона, люди не внесли соответствующую запись, и пропущенное значение можно трактовать как отсутствие балкона;
* `ceiling_height`- доля пропущенных значений в данном столбце составляет 38%, пренебречь этим нельзя. В дальнейшем попробуем заполнить их медианными значениями с группировкой по этажу;
* `airports_nearest` - для заполнения этих пропусков возьмем медианные значения с группировкой по населенному пункту;
* `cityCenters_nearest` - для заполнения этих пропусков возьмем медианные значения с группировкой по населенному пункту;
* `parks_around3000` - для заполнения этих пропусков возьмем медианные значения с группировкой по населенному пункту;
* `ponds_around3000` - для заполнения этих пропусков возьмем медианные значения с группировкой по населенному пункту;
* `days_exposition` - для заполнения пропусков в данном столбце будем отталкиваться от последней даты опубликования объявлений;
* `kitchen_area` - логично предположить, что данные этого столбца - разность общей пложади и жилой;
* `living_area` - положим, что жилая площадь зависит от количества комнат в квартире и заполним пропуски медианными значениями с группировкой по числу комнат;
* `floors_total` - доля пропусков в данном столбце составляет 0,4%. Поскольку пропущенные значения будет не верно заполнять медианными, а процент пропусков ничтожно мал, то можно пренебречь такими строками и удалить их;
* `locality_name` - доля пропусков составляет 0,2%. В дальнейшем эти строки удалим.

#### Столбцы `floors_total`, `locality_name`

В столбцах `floors_total` и `locality_name` доля пропусков наименьшая и составляет 0,4% и 0,2% соответственно. 

Информацию о количестве этажей врядли получится восстановить, единственное от чего может зависеть эта величина - это этаж продаваемой квартиры, однако не исключен тот факт, что помещение может находиться на первом этаже, а всего их - 20. Поэтому справедливо будет удалить строки, в которых есть пропуски в данном столбце.

Наименования населенных пунктов можно было бы заполнить в зависимости от расстояния к различным объектам: парки, водоемы, аэропорт. Однако эти столбцы содержат так много пропусков, что в значительной мере затрудняет такое заполнение. Поэтому строки с пустыми значениями в `locality_name` тоже удалим.

Для этого используем метод `dropna()`:

```python
print('Количество строк до удаления:', data.shape)
data = data.dropna(subset = ['floors_total', 'locality_name'])
print('Количество строк после удаления:', data.shape)
```

Проверим себя:

```python
print('Количество пропусков после удаления строк в столбце floors_total:', data['floors_total'].isna().sum())
print('Количество пропусков после удаления строк в столбце locality_name:', data['locality_name'].isna().sum())
```

Строки с пропусками в столбцах `floors_total` и `locality_name` успешно удалены.

#### Столбец `is_apartment`

Для наглядности обратимся к выборке [первых десяти строк таблицы](#data), в которых можно заметить, что в столбце `is_apartment` содержатся пропуски. Ранее уже было определено, что целесообразно все пропущенные значения заменить на **False**. Используем для этого метод `fillna()` и выведем первые десять строк заново:

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
data.head(10)
```

Проверим себя, посчитав количество пропусков в столбце `is_apartment`:

```python
#подсчет количества пропущенных значений в столбце is_apartment
print('Осталось пропусков:', data['is_apartment'].isna().sum())
```

Замена отработала корректно.

Поэтому справедливо будет заполнить пропуски в данных столбцах их медианными значениями с группировкой по населенному пункту.

Прежде чем начинать считать медианные значения с группировкой по населенным пунктам, попробуем оставить только их названия, используя `str.replace()`. Сперва посмотрим, как они записаны в исходном виде:

```python
#просмотр уникальных значений столбца locality_name
data['locality_name'].unique()
```

В результирующей выборке есть и поселки, и поселки городского типа, и села, и деревни, и прочее. Оставим только названия населенных пунктов:

```python
localities = ['поселок', 'городского', 'станции', 'садоводческое', 'форт', 'имени', 'коттеджный',  'при', 'железнодорожной', 'некоммерческое', 'типа', 'посёлок', 'городской', 'деревня', 'садовое', 'товарищество', 'село']
for locality in localities:
    data['locality_name'] = data['locality_name'].str.replace(locality, '')
    data['locality_name'] = data['locality_name'].str.strip()
data['locality_name'].unique()
```

Типы населенных пунктов удалены, остались лишь навания.

Приступим к заполнению пропущенных значений. Для этого напишем функцию `empty_area(column)`, которая в нужном нам столбце заполнит пропуски медианными значениями этого столбца в зависимости от населенного пункта:

```python
def empty_area(column):
    print('Пропуски до:', data[column].isna().sum())
    for area in data['locality_name'].unique():
        median = data.loc[(data['locality_name'] == area) & (~data[column].isna()), column].median()
        data.loc[(data[column].isna()) & (data['locality_name'] == area), column] = median
    #заполнение оставшихся пропусков, для которых населенный пункт уникален, медианным значением по всему столбцу в целом
    data.loc[data[column].isna(), column] = data[column].median() 
    print('Пропуски после:', data[column].isna().sum())
```

Проверим работоспособность функции на столбце `ponds_nearest`. Сперва выведем первые деять строк из таблицы и посмотрим, есть ли там пропуски:

### Этаж квартиры; варианты — первый, последний, другой

Для того, чтобы это реализовать, необходимо воспользоваться `apply` по всему ряду: для каждого ряда в таблице функция `apply` вызовет функцию `cat_floor` и передаст в нее весь ряд целиком. Результат запишется в новый столбец `cat_floor`:

```python
def cat_floor(value):
    if value['floor'] == 1: return 'первый'
    elif value['floor'] == value['floors_total']: return 'последний'
    return 'другой'
data['cat_floor'] = data.apply(cat_floor, axis = 1)
```

### Анализ редких и выбивающихся значений

Исследуем столбцы `total_area`, `price_per_m2`, `rooms`, `ceiling_height` и `days_exposition` на предмет наличия выбивающихся значений, после чего оставим в таблице все строки за их исключением.

#### Столбец `price_per_m2`

Построим диаграмму размаха:

```python
plt.ylim(0, 250000)
data.boxplot('price_per_m2');
```

```python
len(data)
```

Для удобства создадим функцию, которой воспользуемся при анализе выбивающихся значений и в остальных столбцах. Она определит границы распределения значений, за пределами которых находятся выбросы:

```python
def borders(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3-Q1
    upp_border = Q3 + 1.5 * IQR
    low_border = Q1 - 1.5 * IQR
    rare_upp = data.query(f'{column} > @upp_border')[column].count()
    rare_low = data.query(f'{column} < @low_border')[column].count()
    rare_dolya = round((rare_upp + rare_low) / data[column].count() * 100, 3)
    return (f'Нижняя граница: {low_border}\n'
            f'Верхняя граница: {upp_border}\n'
           f'Процент редких значений: {rare_dolya}')
```

```python
print(borders(data, 'price_per_m2'))
```

Доля редких значений менее 5 %. Из результата выполнения функции видно, что в отфильтрованной таблице должны сохраниться строки, в которых значения в столбце `price_per_m2` находятся в диапазоне от нижней до верхней границы. Все, что выходит за рамки этого диапазона, необходимо отбросить.

Зная это, напишем функцию, которая оставит в таблице соответствующие строки:

```python
def rare_values(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3-Q1
    upp_border = Q3 + 1.5 * IQR
    low_border = Q1 - 1.5 * IQR
    data_filtred = data[(data[column] > low_border) & (data[column] < upp_border)]
    return data_filtred
```

```python
# выполним функцию rare_values() для price_per_m2
data = rare_values(data, 'price_per_m2')
```

Воспользуемся методом `describe()`, чтобы посмотреть, какое минимальное и максимальное значения остались в столбце `price_per_m2`:

```python
data['price_per_m2'].describe()
```

Можем заметить, что минимальное и максимальное значения укладываются в наш диапазон. Функция отработала корректно.
