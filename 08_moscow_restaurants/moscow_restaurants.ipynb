### –û–¥–Ω–æ–æ–±—Ä–∞–∑–Ω–æ–µ –Ω–∞–ø–∏—Å–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π –∑–∞–≤–µ–¥–µ–Ω–∏–π

–£–¥–∞–ª–∏–º –≤—Å–µ–≤–æ–∑–º–æ–∂–Ω—ã–µ –ø—Ä–∏–ø–∏—Å–∫–∏ –≤ –≤–∏–¥–µ "–∫–∞—Ñ–µ", "—Å—Ç–æ–ª–æ–≤–∞—è", "—Ä–µ—Å—Ç–æ—Ä–∞–Ω" –∏ —Ç.–ø., –∞ —Ç–∞–∫–∂–µ –∫–∞–≤—ã—á–∫–∏:

```python
names = ['–∫–∞—Ñ–µ', '–∫–æ—Ñ–µ–π–Ω—è', '–∑–∞–∫—É—Å–æ—á–Ω–∞—è', '–ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏–µ', '–æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è', '–±—ã—Å—Ç—Ä–æ–≥–æ', '–ø–∏—Ç–∞–Ω–∏—è', '—Ä–µ—Å—Ç–æ—Ä–∞–Ω', '—Å—Ç–æ–ª–æ–≤–∞—è', '–±–∞—Ä', '¬´', '¬ª', '-' ]
for name in names:
    rest_data['object_name'] = rest_data['object_name'].str.replace(name, '')
rest_data.head(10)
```

### –°–µ—Ç—å - –Ω–µ —Å–µ—Ç—å?

–ü–æ–ø—Ä–æ–±—É–µ–º –æ—Ç—ã—Å–∫–∞—Ç—å –∑–∞–≤–µ–¥–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –±—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞–ª–∏ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ —Å–ª–µ–¥—É—é—â–∏–º —É—Å–ª–æ–≤–∏—è–º:
1. –í —Å—Ç–æ–ª–±—Ü–µ `chain` –Ω–∞–ø—Ä–æ—Ç–∏–≤ –∑–∞–≤–µ–¥–µ–Ω–∏—è —É–∫–∞–∑–∞–Ω–æ `–¥–∞`.
2. –í—ã–±—Ä–∞–Ω–Ω–æ–µ –∑–∞–≤–µ–¥–µ–Ω–∏–µ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –≤ –ú–æ—Å–∫–≤–µ —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Ä–∞–∑.

–í –Ω–∞–π–¥–µ–Ω–Ω—ã—Ö –∑–∞–≤–µ–¥–µ–Ω–∏—è—Ö –∑–∞–º–µ–Ω–∏–º –ø—Ä–∏–∑–Ω–∞–∫ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç–∏ –∫ —Å–µ—Ç–∏.

```python
# –∏—â–µ–º –Ω–µ —Å–µ—Ç–∏
false_chains = (rest_data.query('chain=="–¥–∞"')
                .groupby('object_name')
                .agg({'object_type': 'count'})
                .query('object_type == 1')
                .reset_index())
# –º–µ–Ω–µ–º —Å—Ç–æ–ª–±–µ—Ü chain
rest_data.loc[rest_data['object_name'].isin(false_chains['object_name']), 'chain'] = '–Ω–µ—Ç'
```

### –°—Ç–∞—Ä–±–∞–∫—Å == Starbucks?

–í –≥–ª–∞–∑–∞ –±—Ä–æ—Å–∞—é—Ç—Å—è –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏—è –æ–±—ä–µ–∫—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞–≤–µ—Ä–Ω—è–∫–∞ —è–≤–ª—è—é—Ç—Å—è –æ–¥–Ω–∏–º –∏ —Ç–µ–º –∂–µ –∑–∞–≤–µ–¥–µ–Ω–∏–µ–º. –ü–µ—Ä–µ–∏–º–µ–Ω—É–µ–º –∏—Ö, —á—Ç–æ–±—ã –∏–∑–±–∞–≤–∏—Ç—å—Å—è –æ—Ç –∑–∞–¥–≤–æ–µ–Ω–∏—è:
* `starbucks coffee`, `starbucks, —Å—Ç–∞—Ä–±–∞–∫—Å –∫–æ—Ñ–µ`, `starbucks –∫–æ—Ñ–µ`, `—Å—Ç–∞—Ä–±–∞–∫—Å`, `—Å—Ç–∞—Ä–±–∞–∫—Å –∫–æ—Ñ–µ`, `–∫–æ—Ñ–µ–º–∞–Ω–∏—è starbucks` ‚Üí **starbucks**;
* `van wok –≤–∞–Ω–≤–æ–∫`, `vanwok –≤–∞–Ω–≤–æ–∫` ‚Üí **vanwok**;
* `–∫—É–ª–∏–Ω–∞—Ä–Ω–æ–µ –±—é—Ä–æ kitchen`, `–∫—É–ª–∏–Ω–∞—Ä–Ω–æ–µ –±—é—Ä–æ –∫–∏—Ç—á–µ–Ω` ‚Üí **kitchen**;
* `–∞–º–µ—Ä–∏–∫–∞–Ω —Å–∏—Ç–∏ –ø–∏—Ü—Ü–∞`, `–∞–º–µ—Ä–∏–∫–∞–Ω —Å–∏—Ç–∏ –ø–∏—Ü—Ü–∞ —Ü–µ—Ö –ø–æ –∏–∑–≥–æ—Ç–æ–≤–ª–µ–Ω–∏—é –ø–∏—Ü—Ü—ã` ‚Üí **–∞–º–µ—Ä–∏–∫–∞–Ω —Å–∏—Ç–∏ –ø–∏—Ü—Ü–∞**;
* `–ø–∞–ø–∞ –¥–∂–æ–Ω—Å –ø–∏—Ü—Ü–∞`, `–ø–∏—Ü–µ—Ä–∏—è –ø–∞–ø–∞ –¥–∂–æ–Ω—Å` ‚Üí **–ø–∞–ø–∞ –¥–∂–æ–Ω—Å**;
* `—Å—Ç–µ–π–∫ —Ö–∞—É—Å –≥—É–¥–º–∞–Ω`, `—Å—Ç–µ–π–∫—Ö–∞—É—Å –≥—É–¥–º–∞–Ω` ‚Üí **–≥—É–¥–º–∞–Ω**.

```python
synonyms = {'starbucks': ['starbucks coffee', 'starbucks –∫–æ—Ñ–µ', '—Å—Ç–∞—Ä–±–∞–∫—Å', '—Å—Ç–∞—Ä–±–∞–∫—Å –∫–æ—Ñ–µ', 'starbucks, —Å—Ç–∞—Ä–±–∞–∫—Å –∫–æ—Ñ–µ', 'starbucks —Å—Ç–∞—Ä–±–∞–∫—Å', '–∫–æ—Ñ–µ–º–∞–Ω–∏—è starbucks'],
           'vanwok': ['van wok –≤–∞–Ω–≤–æ–∫', 'vanwok –≤–∞–Ω–≤–æ–∫'],
           'kitchen': ['–∫—É–ª–∏–Ω–∞—Ä–Ω–æ–µ –±—é—Ä–æ –∫–∏—Ç—á–µ–Ω', '–∫—É–ª–∏–Ω–∞—Ä–Ω–æ–µ –±—é—Ä–æ kitchen'],
           '–∞–º–µ—Ä–∏–∫–∞–Ω —Å–∏—Ç–∏ –ø–∏—Ü—Ü–∞': ['–∞–º–µ—Ä–∏–∫–∞–Ω —Å–∏—Ç–∏ –ø–∏—Ü—Ü–∞ —Ü–µ—Ö –ø–æ –∏–∑–≥–æ—Ç–æ–≤–ª–µ–Ω–∏—é –ø–∏—Ü—Ü—ã'],
           '–ø–∞–ø–∞ –¥–∂–æ–Ω—Å': ['–ø–∞–ø–∞ –¥–∂–æ–Ω—Å –ø–∏—Ü—Ü–∞', '–ø–∏—Ü–µ—Ä–∏—è –ø–∞–ø–∞ –¥–∂–æ–Ω—Å'],
           '–≥—É–¥–º–∞–Ω': ['—Å—Ç–µ–π–∫ —Ö–∞—É—Å –≥—É–¥–º–∞–Ω', '—Å—Ç–µ–π–∫—Ö–∞—É—Å –≥—É–¥–º–∞–Ω']
           }
def check_synonymus(cell):
    for name, syn in synonyms.items():
        if cell in syn: return name
    return cell
rest_data['object_name_new'] = rest_data['object_name'].apply(check_synonymus)

# –ø—Ä–æ–≤–µ—Ä–∫–∞
rest_data[['object_name', 'object_name_new']].query('object_name_new =="–≥—É–¥–º–∞–Ω"')
```

```python
# —É–¥–∞–ª–∏–º —Å—Ç–æ–ª–±–µ—Ü object_name
rest_data = rest_data.drop(columns=['object_name'])

# –ø–µ—Ä–µ–∏–º–µ–Ω—É–µ–º —Å–æ–∑–¥–∞–Ω–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü
rest_data = rest_data.rename(columns={'object_name_new': 'object_name'})

# —É–ø–æ—Ä—è–¥–æ—á–∏–º —Å—Ç–æ–ª–±—Ü—ã –≤ –Ω—É–∂–Ω–æ–º –Ω–∞–º –ø–æ—Ä—è–¥–∫–µ
rest_data = rest_data[['id', 'object_name', 'chain', 'object_type', 'address', 'number']]
rest_data.head()
```

–í –∏—Å—Ö–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ –º–æ–≥–ª–∏ –Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –∑–∞–≤–µ–¥–µ–Ω–∏—è, —É –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ—Ç –Ω–∞–∑–≤–∞–Ω–∏—è. –≠—Ç–æ –º–æ–≥–ª–æ –±—ã—Ç—å –ø—Ä–æ—Å—Ç–æ "–∫–∞—Ñ–µ", "—Ä–µ—Å—Ç–æ—Ä–∞–Ω" –∏ –ø—Ä. –ò–∑-–∑–∞ —Ç–æ–≥–æ, —á—Ç–æ –¥–≤—É–º—è —à–∞–≥–∞–º–∏ —Ä–∞–Ω–µ–µ –º—ã –ø–æ–¥–æ–±–Ω—ã–µ –≤–µ—â–∏ —É–¥–∞–ª–∏–ª–∏ –∏–∑ —Ç–∞–±–ª–∏—Ü—ã, –º–æ–≥–ª–∏ –æ–±—Ä–∞–∑–æ–≤–∞—Ç—å—Å—è –ø—Ä–æ–ø—É—Å–∫–∏. –ü–æ—Å–º–æ—Ç—Ä–∏–º, —Ç–∞–∫ –ª–∏ —ç—Ç–æ –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ:

```python
rest_data.query('object_name == ""')
```

–î–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ, —Ç–∞–∫–∏–µ —Å—Ç—Ä–æ–∫–∏ –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É—é—Ç. –ó–∞–ø–æ–ª–Ω–∏–º –∏—Ö –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏–µ–º `no name`:

```python
def description(object_name):
    if object_name == "":
        return 'no name'
    return object_name
rest_data['object_name'] = rest_data['object_name'].apply(description)

# –ø—Ä–æ–≤–µ—Ä–∏–º
rest_data.query('object_name == "no name"')
```

### –î—É–±–ª–∏–∫–∞—Ç—ã

–ü—Ä–æ–≤–µ—Ä–∏–º, –µ—Å—Ç—å –ª–∏ –≤ —Ç–∞–±–ª–∏—Ü–µ —Å—Ç—Ä–æ–∫–∏, –ø–æ–ª–Ω–æ—Å—Ç—å—é –¥—É–±–ª–∏—Ä—É—é—â–∏–µ –¥—Ä—É–≥ –¥—Ä—É–≥–∞:

–¢–∞–∫–æ–≤—ã–µ –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç.

–ü—Ä–æ–≤–µ—Ä–∏–º –±–∞–∑—É –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –Ω–µ—è–≤–Ω—ã—Ö –¥—É–±–ª–∏–∫–∞—Ç–æ–≤:

```python
# —Å—á–∏—Ç–∞–µ–º —Å—Ç—Ä–æ–∫–∏ –¥–æ —É–¥–∞–ª–µ–Ω–∏—è –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –¥–æ —É–¥–∞–ª–µ–Ω–∏—è –¥—É–±–ª–∏–∫–∞—Ç–æ–≤:', rest_data.shape)

# –∏—â–µ–º –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã
duplicates = rest_data.duplicated(keep = False, subset=['object_name', 'chain', 'object_type', 'address', 'number'])

# –∏—Å–∫–ª—é—á–∞–µ–º –∏—Ö –∏–∑ –æ–±—â–µ–π —Ç–∞–±–ª–∏—Ü—ã
rest_data = rest_data.loc[~duplicates]    

# —Å—á–∏—Ç–∞–µ–º —Å—Ç—Ä–æ–∫–∏ –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –¥—É–±–ª–∏–∫–∞—Ç–æ–≤
print('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –ø–æ—Å–ª–µ —É–¥–∞–ª–µ–Ω–∏—è –¥—É–±–ª–∏–∫–∞—Ç–æ–≤:', rest_data.shape)
```

–ë—ã–ª–æ —É–¥–∞–ª–µ–Ω–æ 156 —Å—Ç—Ä–æ–∫, –¥—É–±–ª–∏—Ä—É—é—â–∏—Ö—Å—è –ø–æ –ø–æ–ª—è–º `object_name`, `chain`, `object_type`, `address`, `number`.

## –ê–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

### –°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –≤–∏–¥–æ–≤ –æ–±—ä–µ–∫—Ç–æ–≤ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É

```python
rest_data.groupby('object_type').agg({'id': 'count'}).sort_values(by='id').plot(kind='barh', legend=False)
plt.suptitle('–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –≤–∏–¥–æ–≤ –æ–±—ä–µ–∫—Ç–æ–≤ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É')
plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤')
plt.ylabel('');
```

–ü–æ –¥–∞–Ω–Ω—ã–º —Ç–∞–±–ª–∏—Ü—ã `rest_data` –≤ –ú–æ—Å–∫–≤–µ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –∫–∞—Ñ–µ. –û–Ω–æ –∏ –ø–æ–Ω—è—Ç–Ω–æ: –≥–æ—Ä–æ–¥ —Ä–∞—Å–ø–æ–ª–∞–≥–∞–µ—Ç –∫ –±—ã—Å—Ç—Ä–æ–º—É —Ç–µ–º–ø—É –∂–∏–∑–Ω–∏, –≥–¥–µ –∏ –ø—Ä–∏–µ–º –ø–∏—â–∏ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –±—ã—Å—Ç—Ä—ã–º, –¥–æ–ª–≥–æ –≤ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞—Ö –Ω–µ –ø–æ—Å–∏–¥–∏—à—å. –ù–∞ –≤—Ç–æ—Ä–æ–º, —Ç—Ä–µ—Ç—å–µ–º –∏ —á–µ—Ç–≤–µ—Ä—Ç–æ–º –º–µ—Å—Ç–µ —Å –Ω–µ–±–æ–ª—å—à–∏–º –æ—Ç—Ä—ã–≤–æ–º –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ —Ä–∞—Å–ø–æ–ª–æ–∂–∏–ª–∏—Å—å —Å—Ç–æ–ª–æ–≤—ã–µ, —Ä–µ—Å—Ç–æ—Ä–∞–Ω—ã –∏ –ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏—è –±—ã—Å—Ç—Ä–æ–≥–æ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è. –ù—É –∞ –º–µ–Ω—å—à–µ –≤—Å–µ–≥–æ –≤ –ú–æ—Å–∫–≤–µ –æ—Ç–¥–µ–ª–æ–≤ –∫—É–¥–∏–Ω–∞—Ä–∏–∏ –≤ –º–∞–≥–∞–∑–Ω–∞—Ö.

### –°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ —Å–µ—Ç–µ–≤—ã—Ö –∏ –Ω–µ—Å–µ—Ç–µ–≤—ã—Ö –∑–∞–≤–µ–¥–µ–Ω–∏–π –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É

```python
report = rest_data.pivot_table(index = 'object_type', columns = 'chain', values = 'id', aggfunc = 'nunique')
report['total'] = report['–¥–∞']+ report['–Ω–µ—Ç']
report = report.sort_values(by='total')
ax = report[['–¥–∞', '–Ω–µ—Ç']].plot(kind='barh', stacked=True)
ax.set_title('–°–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ —Å–µ—Ç–µ–≤—ã—Ö –∏ –Ω–µ—Å–µ—Ç–µ–≤—ã—Ö –∑–∞–≤–µ–¥–µ–Ω–∏–π –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É')
ax.set_xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤')
ax.set_ylabel('');
```

–ù–∞ –≥—Ä–∞—Ñ–∏–∫–µ –≤–∏–¥–Ω–æ, —á—Ç–æ –≤ –æ—Å–Ω–æ–≤–Ω–æ–º –≤ –≥–æ—Ä–æ–¥–µ –ø—Ä–µ–æ–±–ª–∞–¥–∞—é—Ç –Ω–µ—Å–µ—Ç–µ–≤—ã–µ –∑–∞–≤–µ–¥–µ–Ω–∏—è. 

–í–∏–∑—É–∞–ª—å–Ω–æ —Å–µ—Ç–µ–≤–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω–æ –¥–ª—è –ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏–π –±—ã—Å—Ç—Ä–æ–≥–æ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è. –î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —Å–µ–±—è, –ø–æ—Å—á–∏—Ç–∞–µ–º, –∫–∞–∫—É—é –¥–æ–ª—é –∑–∞–Ω–∏–º–∞—é—Ç —Å–µ—Ç–µ–≤—ã–µ –∑–∞–≤–µ–¥–µ–Ω–∏—è –≤ –æ–±—â–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –æ–±—ä–µ–∫—Ç–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤–∏–¥–∞:

```python
report['chain_%'] = round((report['–¥–∞'] / report['total']) * 100, 2)
report.reset_index()[['object_type', 'chain_%']].sort_values(by='chain_%', ascending=False)
```


–î–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ, —Å–µ—Ç–µ–≤–æ–µ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω–µ–Ω–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω–æ –¥–ª—è –ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏–π –±—ã—Å—Ç—Ä–æ–≥–æ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏—è.


### –ß—Ç–æ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω–æ –¥–ª—è —Å–µ—Ç–µ–≤—ã—Ö –∑–∞–≤–µ–¥–µ–Ω–∏–π: –º–Ω–æ–≥–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π —Å –Ω–µ–±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –≤ –∫–∞–∂–¥–æ–º –∏–ª–∏ –º–∞–ª–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π —Å –±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç?

–î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –æ—Ç–≤–µ—Ç–∏—Ç—å –Ω–∞ —ç—Ç–æ—Ç –≤–æ–ø—Ä–æ—Å –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å–æ–≤–µ—Ä—à–∏—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ —à–∞–≥–æ–≤:
1. –í—ã–±—Ä–∞—Ç—å –≤—Å–µ —Å–µ—Ç–µ–≤—ã–µ —Ä–µ—Å—Ç–æ—Ä–∞–Ω—ã.
2. –ü–æ—Å—á–∏—Ç–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π –¥–ª—è –∫–∞–∂–¥–æ–π —Å–µ—Ç–∏, –∞ —Ç–∞–∫–∂–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –¥–ª—è –∫–∞–∂–¥–æ–π —Å–µ—Ç–∏.
3. –†–∞–∑–±–∏—Ç—å –≤—Å–µ —Å–µ—Ç–∏ –Ω–∞ –≥—Ä—É–ø–ø—ã:
* –º–Ω–æ–≥–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π - –º–Ω–æ–≥–æ –º–µ—Å—Ç;
* –º–Ω–æ–≥–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π - –º–∞–ª–æ –º–µ—Å—Ç;
* –º–∞–ª–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π - –º–Ω–æ–≥–æ –º–µ—Å—Ç;
* –º–∞–ª–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π - –º–∞–ª–æ –º–µ—Å—Ç.
4. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –±—É–¥–µ—Ç –≤–∏–¥–Ω–æ –ø—Ä–µ–æ–±–ª–∞–¥–∞–Ω–∏–µ —Ç–æ–π –∏–ª–∏ –∏–Ω–æ–π –≥—Ä—É–ø–ø—ã. 

```python
# —Å–∫–æ–ª—å–∫–æ –≤—Å–µ–≥–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π –≤ –∫–∞–∂–¥–æ–π —Å–µ—Ç–∏, –∞ —Ç–∞–∫–∂–µ —Å—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç
chains = (rest_data.query('chain == "–¥–∞"')
          .groupby('object_name')
          .agg({'id': 'count', 
                'number': 'mean'})
          .round(2)
          .sort_values(by='id', ascending=False).reset_index())
chains.head(10)
```

–î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã —Ä–∞–∑–±–∏—Ç—å –≤—Å–µ —Å–µ—Ç–∏ –Ω–∞ —É—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–Ω—ã–µ —Ä–∞–Ω–µ–µ –≥—Ä—É–ø–ø—ã, –∑–∞ –≥—Ä–∞–Ω–∏—Ü—É –≤–æ–∑—å–º–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ, –Ω–∏–∂–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —á–µ—Ç–≤–µ—Ä—Ç—å –¥–∞–Ω–Ω—ã—Ö –æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–≤–µ–¥–µ–Ω–∏–π / —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç:

```python
# –º–∞–ª–æ/–º–Ω–æ–≥–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π –≤ —Å–µ—Ç–∏
chains['group_rest'] = '–ú–∞–ª–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π'
chains.loc[chains['id'] > np.percentile(chains['id'], 25), 'group_rest'] = '–ú–Ω–æ–≥–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π'

# –º–∞–ª–æ/–º–Ω–æ–≥–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç
chains['group_seat'] = '–ú–∞–ª–æ –º–µ—Å—Ç'
chains.loc[chains['number'] > np.percentile(chains['number'], 25), 'group_seat'] = '–ú–Ω–æ–≥–æ –º–µ—Å—Ç'

# –æ–±—ä–µ–¥–∏–Ω–∏–º —Å—Ç–æ–ª–±—Ü—ã group_rest –∏ group_seat
chains['–≥—Ä—É–ø–ø–∞'] = chains['group_rest'].astype(str) + ' - ' + chains['group_seat'].astype(str)

# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫
plt.figure(figsize=(8, 6))
sns.scatterplot(data = chains, x = 'id', y = 'number', hue = '–≥—Ä—É–ø–ø–∞')
plt.xlabel('–ß–∏—Å–ª–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π')
plt.ylabel('–°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç')
plt.title('–°–µ—Ç–µ–≤—ã–µ –∑–∞–≤–µ–¥–µ–Ω–∏—è');
```

–ò–∑ –≥—Ä–∞—Ñ–∏–∫–∞ —Å–ª–µ–¥—É–µ—Ç, —á—Ç–æ –¥–ª—è —Å–µ—Ç–µ–≤—ã—Ö –∑–∞–≤–µ–¥–µ–Ω–∏–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω–æ –º–∞–ª–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π —Å –±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç.

### –°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤–∏–¥–∞ –æ–±—ä–µ–∫—Ç–∞ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è

–°–ø–µ—Ä–≤–∞ –ø–æ—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤–∏–¥–∞ –æ–±—ä–µ–∫—Ç–∞ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è:

```python
report = (rest_data.pivot_table(index = 'object_type', values = 'number', aggfunc = 'mean')
          .round()
          .sort_values(by='number', ascending=False))
report
```

–ü–æ—Å—Ç—Ä–æ–∏–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –≥—Ä–∞—Ñ–∏–∫–æ–≤: —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –≤ —Å—Ä–µ–¥–Ω–µ–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤–∏–¥–∞ –æ–±—ä–µ–∫—Ç–∞ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è –∏ –≤ —Ä–∞–∑–±–∏–≤–∫–µ –ø–æ —Å–µ—Ç–µ–≤—ã–º / –Ω–µ—Å–µ—Ç–µ–≤—ã–º –∑–∞–≤–µ–¥–µ–Ω–∏—è–º.

```python
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–µ–¥–Ω–∏—Ö –º–µ—Å—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤–∏–¥–∞ –æ–±—ä–µ–∫—Ç–∞ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è –≤ —Ü–µ–ª–æ–º
report = rest_data.pivot_table(index = 'object_type', values = 'number', aggfunc = 'mean')
report = report.sort_values(by='number')
ax = report.plot(kind='barh', legend=False)
ax.set_title('–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–µ—Å—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤–∏–¥–∞ –æ–±—ä–µ–∫—Ç–∞ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è')
ax.set_xlabel('–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç')
ax.set_ylabel('');
```

–ò–∑ –≥—Ä–∞—Ñ–∏–∫–∞ —Å–ª–µ–¥—É–µ—Ç, —á—Ç–æ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –≤ —Å—Ç–æ–ª–æ–≤—ã—Ö –∏ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞—Ö. –ó–∞—Ç–µ–º —Å –Ω–µ–±–æ–ª—å—à–∏–º –æ—Ç—Ä—ã–≤–æ–º –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ –∏–¥—É—Ç –±—É—Ñ–µ—Ç—ã, –±–∞—Ä—ã –∏ –∫–∞—Ñ–µ. –°–æ–≤—Å–µ–º –º–∞–ª–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Ç –≤ –æ—Ç–¥–µ–ª–∞—Ö –∫—É–ª–∏–Ω–∞—Ä–∏–∏ –≤ –º–∞–≥–∞–∑–∏–Ω–∞—Ö.

```python
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ä–µ–¥–Ω–∏—Ö –º–µ—Å—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤–∏–¥–∞ –æ–±—ä–µ–∫—Ç–∞ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è –≤ —Ä–∞–∑–±–∏–≤–∫–µ –Ω–∞ —Å–µ—Ç–µ–≤—ã–µ / –Ω–µ —Å–µ—Ç–µ–≤—ã–µ
report = rest_data.pivot_table(index = 'object_type', columns = 'chain', values = 'number', aggfunc = 'mean')
report['total'] = report['–¥–∞']+ report['–Ω–µ—Ç']
report = report.sort_values(by='total')
ax = report[['–¥–∞', '–Ω–µ—Ç']].plot(kind='barh', stacked=True)
ax.set_title('–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–µ—Å—Ç –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤–∏–¥–∞ –æ–±—ä–µ–∫—Ç–∞ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è (—Å–µ—Ç–µ–≤—ã–µ / –Ω–µ —Å–µ—Ç–µ–≤—ã–µ)')
ax.set_xlabel('–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç')
ax.set_ylabel('');
```

–ß—Ç–æ –º—ã –≤–∏–¥–∏–º:
1. –ë–æ–ª—å—à–µ –≤—Å–µ–≥–æ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –≤ –Ω–µ—Å–µ—Ç–µ–≤—ã—Ö —Å—Ç–æ–ª–æ–≤—ã—Ö –∏ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞—Ö. 
2. –ü—Ä–∏–º–µ—á–∞—Ç–µ–ª—å–Ω–æ, —á—Ç–æ –≤ —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞—Ö —Ä–∞–∑–Ω–∏—Ü—ã –º–µ–∂–¥—É –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –≤ —Å–µ—Ç–µ–≤–æ–º –∏ –Ω–µ—Å–µ—Ç–µ–≤–æ–º —Ç–∏–ø–∞—Ö –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –Ω–µ—Ç.
3. –¢–∞–∫–∂–µ –ª—å–≤–∏–Ω—É—é –¥–æ–ª—é –æ—Ç –æ–±—â–µ–≥–æ —á–∏—Å–ª–∞ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –≤ –±—É—Ñ–µ—Ç–∞—Ö –∑–∞–Ω–∏–º–∞—é—Ç –Ω–µ—Å–µ—Ç–µ–≤—ã–µ –∑–∞–≤–µ–¥–µ–Ω–∏—è.

### –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ–± —É–ª–∏—Ü–µ

–í—ã–¥–µ–ª–∏–º –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü `streetname` –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ–± —É–ª–∏—Ü–µ –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ `address`:

```python
def get_streetname(cell):
    for addr_part in cell.split(','):
        for s_type in ['—É–ª–∏—Ü–∞', '–±—É–ª—å–≤–∞—Ä', '–ø–µ—Ä–µ—É–ª–æ–∫', '—à–æ—Å—Å–µ', '–ø—Ä–æ—Å–ø–µ–∫—Ç',]:
            if s_type in addr_part:
                return addr_part
    return cell.split(',')[1]
rest_data['streetname'] = rest_data['address'].apply(get_streetname)
rest_data['streetname'] = rest_data['streetname'].str.strip()
# –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
rest_data.head(10)
```

### –¢–æ–ø-10 —É–ª–∏—Ü –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –æ–±—ä–µ–∫—Ç–æ–≤ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è

–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ —Ç–æ–ø-10 —É–ª–∏—Ü –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –æ–±—ä–µ–∫—Ç–æ–≤ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è:

```python
street_10 = (rest_data.pivot_table(index = 'streetname', values = 'id', aggfunc = 'count')
             .rename(columns={'id': 'objects_count'})
             .sort_values(by='objects_count', ascending=False)
             .head(10))
ax = street_10.plot(kind='barh', legend=False)
ax.set_title('–¢–æ–ø-10 —É–ª–∏—Ü –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –æ–±—ä–µ–∫—Ç–æ–≤ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è')
ax.set_xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–≤–µ–¥–µ–Ω–∏–π')
ax.set_ylabel('');
```

–í –≤—ã–±–æ—Ä–∫—É –ø–æ–ø–∞–ª –≥–æ—Ä–æ–¥ –ó–µ–ª–µ–Ω–æ–≥—Ä–∞–¥. –ü–æ—Å–∫–æ–ª—å–∫—É –º–∞–º –≤—Å–µ —Ä–∞–≤–Ω–æ –Ω–µ—Ç —É–ª–∏—Ü –∫–∞–∫ —Ç–∞–∫–æ–≤—ã—Ö - –≤ –≥–æ—Ä–æ–¥–µ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω–æ–µ –¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –º–∏–∫—Ä–æ—Ä–∞–π–æ–Ω—ã, –æ–Ω –±—É–¥–µ—Ç –≤–æ—Å–ø—Ä–∏–Ω–∏–º–∞—Ç—å—Å—è –∫–∞–∫ –æ–¥–Ω–∞ —É–ª–∏—Ü–∞. –í –∫–∞–∫–æ–º-—Ç–æ —Å–º—ã—Å–ª–µ —ç—Ç–æ –º–æ–∂–µ—Ç –∏—Å–∫–∞–∑–∏—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∞–Ω–∞–ª–∏–∞–∑. –ò—Å–∫–ª—é—á–∏–º –µ–≥–æ:

```python
street_10 = (rest_data.query('streetname != "–≥–æ—Ä–æ–¥ –ó–µ–ª–µ–Ω–æ–≥—Ä–∞–¥"').pivot_table(index = 'streetname', values = 'id', aggfunc = 'count')
             .rename(columns={'id': 'object_count'})
             .sort_values(by='object_count', ascending=False)
             .head(10))
street_10
```

–†–∞–Ω–µ–µ –Ω–∞–º–∏ –±—ã–ª –≤—Å—Ç—Ä–æ–µ–Ω —Å—Ç–æ—Ä–æ–Ω–Ω–∏–π —Ñ–∞–π–ª —Å —É–ª–∏—Ü–∞–º–∏ –ú–æ—Å–∫–≤—ã. –û—Ç–∫—Ä–æ–µ–º –µ–≥–æ:

```python
streets.head()
```

–°—Ç–æ–ª–±—Ü—ã `areaid` –∏ `okrug` –Ω–∞–º –Ω–µ –ø–æ–Ω–∞–¥–æ–±—è—Ç—Å—è. –û—Ç—Å–µ—á–µ–º –∏—Ö:

```python
streets = streets[['streetname', 'area']]
```

–î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã —É–∑–Ω–∞—Ç—å, –∫–∞–∫–æ–º—É —Ä–∞–π–æ–Ω—É –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—Ç —É–ª–∏—Ü—ã –∏–∑ —Ç–æ–ø-10, –æ–±—ä–µ–¥–∏–Ω–∏–º `streets` –∏ `street_10`:

```python
street_10_area = street_10.merge(streets, on = 'streetname', how = 'left')
street_10_area
```

–ü–æ—Å–º–æ—Ç—Ä–∏–º, —á–µ—Ä–µ–∑—á–µ—Ä–µ–∑ —Å–∫–æ–ª—å–∫–æ —Ä–∞–π–æ–Ω–æ–≤ –ø—Ä–æ—Ö–æ–¥—è—Ç —Ç–æ–ø-10 —É–ª–∏—Ü:

```python
(street_10_area.groupby('streetname')
 .agg({'area': 'count'})
 .reset_index()
 .rename(columns={'area': '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–π–æ–Ω–æ–≤', 'streetname': '–£–ª–∏—Ü–∞'})
 .sort_values(by='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–π–æ–Ω–æ–≤', ascending=False))
```

–í–∞—Ä—à–∞–≤—Å–∫–æ–π —à–æ—Å—Å–µ –∏ –õ–µ–Ω–∏–Ω—Å–∫–∏–π –ø—Ä–æ—Å–ø–µ–∫—Ç –∑–∞—Ç—Ä–∞–≥–∏–≤–∞—é—Ç –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ —Ä–∞–π–æ–Ω–æ–≤ - 8. –ê –≤–æ—Ç –ü—Ä–µ—Å–Ω–µ–Ω—Å–∫–∞—è –Ω–∞–±–µ—Ä–µ–∂–Ω–∞—è —Ä–∞—Å–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö –æ–¥–Ω–æ–≥–æ —Ä–∞–π–æ–Ω–∞.

### –ß–∏—Å–ª–æ —É–ª–∏—Ü —Å –æ–¥–Ω–∏–º –æ–±—ä–µ–∫—Ç–æ–º –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è

–û—Ç–±–µ—Ä–µ–º —É–ª–∏—Ü—ã, –Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö —Ä–∞—Å–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è –ª–∏—à—å –æ–¥–Ω–æ –∑–∞–≤–µ–¥–µ–Ω–∏–µ:

```python
street_one_object = (rest_data.groupby('streetname')
                     .agg({'object_type': 'count'})
                     .rename(columns={'object_type': 'objects_count'})
                     .query('objects_count == 1'))
street_one_object
```

574 —É–ª–∏—Ü—ã –∏–º–µ—é—Ç —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –æ–±—ä–µ–∫—Ç –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è.

–ü–æ—Å–º–æ—Ç—Ä–∏–º, –≤ –∫–∞–∫–∏—Ö —Ä–∞–π–æ–Ω–∞—Ö —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω—ã –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —É–ª–∏—Ü—ã:

```python
street_one_object_area = street_one_object.merge(streets, on = 'streetname', how = 'left')
street_one_object_area.head(10)
```

–í 6 —Å—Ç—Ä–æ–∫–µ –≤ —Å—Ç–æ–ª–±—Ü–µ `area` –∏–º–µ–µ—Ç—Å—è –ø—Ä–æ–ø—É—Å–∫. –ù–µ –∏—Å–∫–ª—é—á–µ–Ω–æ, —á—Ç–æ –æ–Ω –Ω–µ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–π. –ü–æ—Å–º–æ—Ç—Ä–∏–º, —Å–∫–æ–ª—å–∫–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Ç–∞–±–ª–∏—Ü–µ `street_one_object_area`:

```python
print('–í —Å—Ç–æ–ª–±—Ü–µ area —Ç–∞–±–ª–∏—Ü—ã street_one_object_area —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è:', street_one_object_area['area'].isna().sum(), '–ø—Ä–æ–ø—É—Å–∫–∞(–æ–≤).')
```

–ó–∞–ø–æ–ª–Ω–∏–º –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ñ—Ä–∞–∑–æ–π `–Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω`:

–í—ã–≤–µ–¥–µ–º —Ç–æ–ø-10 —Ä–∞–π–æ–Ω–æ–≤, –≤ –∫–æ—Ç–æ—Ä—ã—Ö —Å–æ—Å—Ä–µ–¥–æ—Ç–æ—á–µ–Ω—ã —É–ª–∏—Ü—ã —Å –æ–¥–Ω–∏–º –æ–±—ä–µ–∫—Ç–æ–º –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è:

```python
(street_one_object_area.groupby('area')
 .agg({'streetname': 'count'})
 .rename(columns={'streetname': 'street_count'})
 .sort_values(by='street_count', ascending=False)
 .head(10))
```

### –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–æ—Å–∞–¥–æ—á–Ω—ã—Ö –º–µ—Å—Ç –¥–ª—è —É–ª–∏—Ü —Å –±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –æ–±—ä–µ–∫—Ç–æ–≤ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–∏—Ç–∞–Ω–∏—è

–ò—Å–ø–æ–ª—å–∑—É–µ–º —Ä–∞–Ω–µ–µ —Å–æ–∑–¥–∞–Ω–Ω—ã–π —Å–ø—Ä–∞–≤–æ—á–Ω–∏–∫ `street_10`. –°–ø–µ—Ä–≤–∞ –Ω–∞ –æ—Å–Ω–æ–≤–µ —Ç–∞–±–ª–∏—Ü—ã `rest_data` —Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º `street_seat_number`, –≤ –∫–æ—Ç–æ—Ä–æ–º –æ—Å—Ç–∞–Ω—É—Ç—Å—è –¥–∞–Ω–Ω—ã–µ —Ç–æ–ª—å–∫–æ –ø–æ –∑–∞–≤–µ–¥–µ–Ω–∏—è–º, –∫–æ—Ç–æ—Ä—ã–µ —Ä–∞—Å–ø–æ–ª–∞–≥–∞—é—Ç—Å—è –Ω–∞ —É–ª–∏—Ü–∞—Ö, —Å–æ–¥–µ—Ä–∂–∞—â–∏—Ö—Å—è –≤ `street_10`:
